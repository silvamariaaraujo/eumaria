```python
dados["Qual é o maior planeta do Sistema Solar?"].hist()
```

```python
dados["Em que continente fica o Egito?"].hist()
```

## Resumo
🌍 Resumo do histograma — “Qual é o maior planeta do Sistema Solar?”

O histograma dessa pergunta costuma mostrar uma concentração muito grande de respostas em Júpiter, já que ele é amplamente conhecido como o maior planeta.
Assim, o gráfico tende a ter uma barra muito alta para “Júpiter” e barras pequenas — ou praticamente inexistentes — para respostas incorretas, como Saturno, Netuno, Urano, Terra, etc.
Em outras palavras, o padrão geral indica que a maioria das pessoas respondeu corretamente.

🌍 Resumo do histograma — “Em que continente fica o Egito?”

O histograma dessa pergunta normalmente apresenta uma concentração elevada de respostas em África, pois é a resposta correta e bastante conhecida.
Barras menores podem aparecer para respostas incorretas, como Ásia ou Europa, indicando alguns erros pontuais.
O padrão geral sugere que a maior parte dos respondentes identifica corretamente o continente, mas há uma pequena parcela que ainda confunde a localização geográfica do país.

## Resumo

🇧🇷 Resumo — “Qual é o idioma oficial do Brasil?”

O resultado dessa pergunta costuma mostrar uma grande concentração de respostas em “Português”, que é amplamente reconhecido como o idioma oficial do país.
As respostas incorretas, como Espanhol, Inglês ou Francês, geralmente aparecem em números muito pequenos ou quase inexistem.
O padrão geral indica que a maioria absoluta dos respondentes acertou, demonstrando conhecimento básico sobre a língua oficial do Brasil.

📚 Resumo — “Quem escreveu Dom Casmurro?”

O histograma dessa pergunta normalmente revela uma forte predominância de respostas em “Machado de Assis”, que é o autor amplamente conhecido do romance.
Outros nomes incorretos, como José de Alencar, Lima Barreto ou Clarice Lispector, tendem a aparecer ocasionalmente e em proporções bem menores.
Assim, o padrão sugere que a maior parte das pessoas respondeu corretamente, mas ainda existe um pequeno grupo que demonstra incerteza sobre literatura brasileira clássica.
